(You can infer this by looking at how the yellow dots cluster at higher GrLivArea values, while the blue dots tend to correspond to lower GrLivArea.) This signifies that the owners of sold houses who have a higher budget for larger living areas, tend to also have the budget for higher quality materials for the interior design and finish of the house.

Question 4: Marginal Histogram Plot [2 Marks]
---

<br>
<div class="alert alert-block alert-info">

<b>Background:</b><br>
- <i>"How does the age of a house relate its sale prices," </i> you wonder. You then realise there is a YearBuilt column in the dataset, which you find perfect for answering this question.<br>
- To leverage on the multi-dimensional visualisation capacity of the new plotly express library you learnt, you decide to add in another feature to answer a second question: <i>"What about the foundation type that a house is built on?"</i> <br>

<b>User story:</b><br>
- You have been given a dataframe <code>q4_df</code> that lists houses with one of 3 foundation types - Poured Concrete (PConc), Cinder Blocks (CBlock), and Brick & Tiles (BrkTil).<br>
- Using <code>q4_df</code> and <b>plotly.express</b>, generate a scatterplot of `SalePrice` against `YearBuilt`. <br>
- Next, add a <b>marginal histogram plot to the y-axis</b> to visualise the distribution of sale prices for different `Foundation` types.<br>
- Expected output:<br><br>

<div style="margin-left:30px;background:white;">


    
    
</div>


```python
foundation_types = ['PConc','CBlock','BrkTil']
q4_df = housing_prices_df[housing_prices_df.Foundation.isin(foundation_types)]

## Write your code below

fig = px.scatter(q4_df, x = "YearBuilt", y ="SalePrice", color="Foundation", marginal_y="histogram")
fig.show()
```

<div class="alert alert-block alert-warning">
Looking at the foundation type of houses built across the years, we can see that <b>most Brick-&-Tile houses were built in the pre-1940s</b>. This is <b>followed by Cinder Block houses</b> mostly built in the mid 20th century, and <b>modern houses these days tend to be built on Poured Concrete</b>, which is much stronger and easier to waterproof.<br>
    
This likely accounts for the higher prices that newly-built poured concrete houses can fetch -- on the upwards of half a million and more!

Question 5: Boxplots [2 Marks]
---

<br>
<div class="alert alert-block alert-info">

<b>Background:</b><br>
- As a potential house buyer in Iowa, one might wonder what is the price to pay for houses with better interior & exterior finishes.<br>
- <i>"What is the distribution of sale prices for houses of various overall & exterior quality?"</i> You set out to use boxplots to investigate this question.<br><br>

<b>User story:</b><br>
- Using <code>q5_df</code> and <b>plotly.express, generate a boxplot</b> that displays the <code>SalePrice</code> distribution for various Overall Quality and Exterior Quality scores.<br>
- You should use <code>OverallQual</code> in the x-axis and <code>ExterQual</code> for color.<br>
- Make sure to provide a <b>descriptive title</b> to the boxplot.<br>
- Expected output:<br><br>

<div style="margin-left:30px;background:white;">

    
    
</div>


```python
high_overall_qual = housing_prices_df.OverallQual >= 6
good_exterior_qual = housing_prices_df.ExterQual.isin(['Gd','Ex']) # Good or Excellent
q5_df = housing_prices_df[high_overall_qual & good_exterior_qual]

## Write your code below

fig = px.box(q5_df, x="OverallQual", y="SalePrice", color="ExterQual", title="Sale Price vs Exterior and Overall Quality")
fig.show()


```

<div class="alert alert-block alert-warning">
    
In general, we find from this chart that the <b>higher the overall and exterior material quality of the houses, the higher the sale prices.</b><br>

Something peculiar about this chart that for houses with Overall Quality score of 10, those houses with more excellent exterior quality tend to fetch significantly lower sale prices across the board. A further investigation into these houses tell us there are very few 10-score houses with "Gd" exterior quality rating, which could have resulted in the unusual upward skew for that boxplot.
