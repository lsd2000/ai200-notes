```python
# Hence, the following:
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer, make_column_selector

colsToEncodeOrdinal = [
    "Month",
    "DayOfWeek",
    "VehiclePrice",
    "Days_Policy_Accident",
    "Days_Policy_Claim",
    "PastNumberOfClaims",
    "AgeOfVehicle",
    "AgeOfPolicyHolder",
    "NumberOfSuppliments",
    "AddressChange_Claim",
    "NumberOfCars",
]

colsToEncodeOHE = [
    "Make",
    "AccidentArea",
    "DayOfWeekClaimed",
    "MonthClaimed",
    "Sex",
    "MaritalStatus",
    "Fault",
    "VehicleCategory",
    "BasePolicy",
    "PoliceReportFiled",
    "WitnessPresent",
    "AgentType",
    # "PolicyType",  # usually drop because it duplicates VehicleCategory + BasePolicy
]

ord_categories = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "DayOfWeek": ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
    "VehiclePrice": ["less than 20000","20000 to 29000","30000 to 39000",
                     "40000 to 59000","60000 to 69000","more than 69000"],
    "Days_Policy_Accident": ["none","1 to 7","8 to 15","15 to 30","more than 30"],
    "Days_Policy_Claim": ["8 to 15","15 to 30","more than 30"],
    "PastNumberOfClaims": ["none","1","2 to 4","more than 4"],
    "AgeOfVehicle": ["new","2 years","3 years","4 years","5 years","6 years","7 years","more than 7"],
    "AgeOfPolicyHolder": ["16 to 17","18 to 20","21 to 25","26 to 30","31 to 35",
                          "36 to 40","41 to 50","51 to 65","over 65"],
    "NumberOfSuppliments": ["none","1 to 2","3 to 5","more than 5"],
    "AddressChange_Claim": ["under 6 months","1 year","2 to 3 years","4 to 8 years","no change"],
    "NumberOfCars": ["1 vehicle","2 vehicles","3 to 4","5 to 8","more than 8"],
}

ord_categories_list = [ord_categories[c] for c in colsToEncodeOrdinal]

# hence, creating the pipeline for preprocessors to be used later

ord_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder(
        categories=ord_categories_list,
        handle_unknown="use_encoded_value",
        unknown_value=-1,
    ),
)

ohe_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False),
)

num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler(with_mean=True, with_std=True),
)

pre = ColumnTransformer(
    transformers=[
        ("ord", ord_pipeline, colsToEncodeOrdinal),
        ("ohe", ohe_pipeline, colsToEncodeOHE),
        ("num", num_pipeline, make_column_selector(dtype_include=np.number)),
    ],
    remainder="drop",
)

```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.pipeline import Pipeline

X = df.drop(columns = ["FraudFound_P"])
y = df["FraudFound_P"].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify = y, random_state=42
)

# Handle class imbalance
pos = (y_train == 1).sum()
neg = (y_train == 0).sum()
spw = neg / max(pos, 1)

pipe = Pipeline([
    ("pre", pre),
    ("clf", XGBClassifier(
        tree_method="hist",          # fast, good default
        random_state=42,
        eval_metric="logloss",       # silence warnings
        n_jobs=-1,
        scale_pos_weight=spw         # imbalance handling
    ))
])

param_grid = {
    "clf__n_estimators":   [300, 600],
    "clf__max_depth":      [3, 5, 7],
    "clf__learning_rate":  [0.05, 0.1],
    "clf__subsample":      [0.7, 1.0],
    "clf__colsample_bytree":[0.7, 1.0],
    "clf__reg_lambda":     [1.0, 5.0],
    "clf__gamma":          [0, 1]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
gs = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=cv,
    scoring="roc_auc",     # use PR AUC if base rate is very low: 'average_precision'
    n_jobs=-1,
    verbose=1,
    refit=True             # refit on full train with best params
)

gs.fit(X_train, y_train)



```

```python
from sklearn.metrics import (
    roc_auc_score, classification_report,
    precision_recall_curve, roc_curve
)

#maximise recall 

y_prob = gs.predict_proba(X_test)[:, 1]
print("Test ROC AUC:", roc_auc_score(y_test, y_prob))

# --------------------------------------------
# 1) Max-Recall threshold (choose the largest Ï„ that attains max recall)
# --------------------------------------------
p, r, th = precision_recall_curve(y_test, y_prob)
# r, p have length n; th has length n-1 and aligns with r[1:], p[1:]
r_max = r.max()
idxs = np.where(r[1:] == r_max)[0]
tau_recall = th[idxs[-1]] if len(idxs) else th[0]   # largest Ï„ achieving r_max

y_pred_recall = (y_prob >= tau_recall).astype(int)
print(f"[Max-Recall] threshold = {tau_recall:.6f}, recall = {r_max:.3f}")

target_names = ["Not Fraud", "Fraud"]
unique_labels = [0, 1]

report_recall = classification_report(
    y_test, y_pred_recall,
    target_names=target_names, labels=unique_labels, output_dict=True
)
report_recall_df = pd.DataFrame(report_recall).transpose()
print(report_recall_df)
```

```python
# --------------------------------------------
# 2) Youdenâ€™s J threshold (max TPR - FPR on ROC)
# --------------------------------------------
fpr, tpr, thr = roc_curve(y_test, y_prob)
j_scores = tpr - fpr
j_idx = np.argmax(j_scores)
tau_youden = thr[j_idx]

y_pred_youden = (y_prob >= tau_youden).astype(int)
print(f"[Youden] threshold = {tau_youden:.6f}, TPR = {tpr[j_idx]:.3f}, FPR = {fpr[j_idx]:.3f}")

report_youden = classification_report(
    y_test, y_pred_youden,
    target_names=target_names, labels=unique_labels, output_dict=True
)
report_youden_df = pd.DataFrame(report_youden).transpose()
print(report_youden_df)
```

```python
# --------------------------------------------
# 3) F1-max threshold (balances precision & recall)
# --------------------------------------------
# thresholds 'th_f1' align with p_f1[1:], r_f1[1:]
p_f1, r_f1, th_f1 = precision_recall_curve(y_test, y_prob)
f1_scores = 2 * p_f1 * r_f1 / (p_f1 + r_f1 + 1e-9)

best_idx_f1 = np.nanargmax(f1_scores[1:])     # ignore the first point (no threshold)
tau_f1 = th_f1[best_idx_f1]

y_pred_f1 = (y_prob >= tau_f1).astype(int)
print(f"[F1-max] threshold = {tau_f1:.6f}, "
      f"F1 = {f1_scores[1:][best_idx_f1]:.3f}, "
      f"precision = {p_f1[1:][best_idx_f1]:.3f}, "
      f"recall = {r_f1[1:][best_idx_f1]:.3f}")

report_f1 = classification_report(
    y_test, y_pred_f1,
    target_names=target_names, labels=unique_labels, output_dict=True
)
report_f1_df = pd.DataFrame(report_f1).transpose()
print(report_f1_df)
```

```python
# --------------------------------------------
# 4) Max recall subject to a precision floor
# --------------------------------------------
from sklearn.metrics import precision_recall_curve

PREC_FLOOR = 0.20  # <-- set your minimum acceptable precision here

p, r, th = precision_recall_curve(y_test, y_prob)  # th aligns with p[1:], r[1:]

# indices where precision meets the floor
ok = np.where(p[1:] >= PREC_FLOOR)[0]

if len(ok):
    # among those, pick the index with the highest recall
    best_idx = ok[np.argmax(r[1:][ok])]
    tau_hiR_precFloor = th[best_idx]
    prec_best = p[1:][best_idx]
    rec_best  = r[1:][best_idx]
else:
    # fallback: if no threshold satisfies the precision floor, pick the highest precision point
    warn_msg = f"[WARN] No threshold meets precision >= {PREC_FLOOR:.2f}. Falling back to highest-precision point."
    print(warn_msg)
    best_idx = np.argmax(p[1:])
    tau_hiR_precFloor = th[best_idx]
    prec_best = p[1:][best_idx]
    rec_best  = r[1:][best_idx]

y_pred_hiR_precFloor = (y_prob >= tau_hiR_precFloor).astype(int)
print(f"[Hi-Recall @ Precisionâ‰¥{PREC_FLOOR:.2f}] "
      f"threshold = {tau_hiR_precFloor:.6f}, precision = {prec_best:.3f}, recall = {rec_best:.3f}")

report_hiR_precFloor = classification_report(
    y_test, y_pred_hiR_precFloor,
    target_names=target_names, labels=unique_labels, output_dict=True
)
report_hiR_precFloor_df = pd.DataFrame(report_hiR_precFloor).transpose()
print(report_hiR_precFloor_df)
```

```python
# y_prob, y_test, target_names, unique_labels already defined above

RECALL_MIN = 0.85   # <-- set your recall target here (e.g., 0.80â€“0.90)
BETA = 2.0          # F2 fallback if target recall is not achievable

p, r, th = precision_recall_curve(y_test, y_prob)   # th aligns with p[1:], r[1:]

# -------- Try: meet recall target, maximize precision among feasible thresholds --------
feasible = np.where(r[1:] >= RECALL_MIN)[0]

if len(feasible):
    # among feasible thresholds, choose the one with highest precision
    best_feasible_idx = feasible[np.argmax(p[1:][feasible])]
    tau = th[best_feasible_idx]
    chosen = f"Recallâ‰¥{RECALL_MIN:.2f} (best precision among feasible)"
else:
    # -------- Fallback: maximize F_beta (recall-weighted) --------
    f_beta = (1 + BETA**2) * p * r / (BETA**2 * p + r + 1e-9)
    best_idx = np.nanargmax(f_beta[1:])   # ignore first point (no threshold)
    tau = th[best_idx]
    chosen = f"F{int(BETA)}-max fallback (recall-weighted)"

# Apply threshold and show report
y_pred = (y_prob >= tau).astype(int)
print(f"[Chosen] {chosen}  |  threshold = {tau:.6f}")

report = classification_report(
    y_test, y_pred,
    target_names=target_names, labels=unique_labels, output_dict=True
)
report_df = pd.DataFrame(report).transpose()
print(report_df)
```
