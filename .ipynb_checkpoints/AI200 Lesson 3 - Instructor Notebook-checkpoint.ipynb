Imagine you are working for a roboadvisor company (i.e Stashaway) and you were tasked to use data to examine what advertising channel will lead to the best outcome in terms of sales.

<b>3b.i) Read data</b>

First we read the advertising data for financial products into pandas dataframe

```python
import pandas as pd
bank_advertising_df = pd.read_csv('data/advertising.csv')
bank_advertising_df
```

Before we proceed, we apply <code>.describe()</code> on the data to get a better understanding of the data we are dealing with

Next, we create a simple linear plot of each of the features against the outcome variable. This allows us to understand if the features have a linear reletionaship with the outcome variable.

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Here we tell plotly the size of the subplot
fig = make_subplots(rows=1, cols=3)

for index, colname in enumerate(['Radio', 'Newspaper', 'TV']):
    
    # Add subplots
    fig.add_trace(
        go.Scatter(x=bank_advertising_df[colname], y=bank_advertising_df["Finance_Product_Sales"], 
                   mode="markers", name=colname),
        row=1, col=index+1  # column indexes are: 1, 2, 3
    )

    # Update xaxis properties
    fig.update_xaxes(title_text=f"{colname} Spending", row=1, col=index+1)

    # Update yaxis properties
    fig.update_yaxes(title_text="Sales", row=1, col=index+1)

# Update title and height
fig.update_layout(height=500, width=1000, title_text="Sales vs Advertising Channels")
fig.show()
```

<b>3b.iii) Train Model</b>

Now we can begin to create the linear multiple regression model. To do this, we make use of a library call - <b>scikit-learn (aka Sklearn)</b>. You can find Sklearn's documentations on linear regression in the link below:

- https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

We start by training the model on the features <b>TV</b>, <b>Radio</b> and <b>Newspaper</b>. To do so:
- We first create a linear regression object using the <code>LinearRegression()</code> function
- Thereafter we fit the model on the <b>training data</b> using the <code>fit()</code> function

```python
from sklearn import linear_model

# Create linear regression object
linear_reg = linear_model.LinearRegression()

# Split data into predictors X and output Y
features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features] # Print x out if you don't understand what data it contains
y = bank_advertising_df['Finance_Product_Sales']

# Initialise and fit model linear multiple regression model using TV, Radio and Newspaper as predictors
linear_reg_model = linear_reg.fit(X, y)
```

To see the model's coefficient, we can use <code>.intercept_</code> and <code>.coef_</code>, and we can see that our model is actually:<br>

<h4>Sales = 4.63 + (0.054 x TV) + (0.107 x Radio) + (0.00033 x Newspaper)</h4><br>

```python
# To see the model's coefficient:
print(f'alpha = {linear_reg_model.intercept_}')
print(f'betas = {linear_reg_model.coef_}')
```

<b>3b.iv) Generate predictions</b>

To generate predictions on <b>data beyond the training dataset</b>, we just have to pass the <code>predict()</code> function a list of the x-values for which we want to generate predictions for. Imagine your marketing team is proposing a new marketing budget composition for next month's campaign and wishes to run the numbers through your model to see if it will achieve the desired outcome. 

Given a budget of TV=292.9, Radio=28.4 and Newspaper=43.2:

```python
# TV=292.9, Radio=28.4 and Newspaper=43.2

new_X = [[292.9, 28.4, 43.2]]
predicted_sales = linear_reg_model.predict(new_X)

print(predicted_sales)
```

<b>3b.v) Calculate RMSE</b>

To calculate the <b>RMSE</b> of your prediction:
- you need to <code>import mean_squared_error</code>
- pass in the true outcome and your predicted outcome into <code>mean_squared_error()</code>. You also need to pass in the parameter <code>squared=False</code> to get the RMSE

```python
from sklearn.metrics import mean_squared_error
import math

actual_sales = [21.5] # actual sales is 21.5
RMSE = mean_squared_error(actual_sales, predicted_sales, squared=False) # squared is True by default

print(RMSE)
```

<b>3b.vi) Validation using Train-Test Split</b>

We covered that train-test split is a technique that estimates a models test error by leaving  out a subset of provided data during training / fitting process and using this subset as a test dataset. 

To do so:
- we can use the <code>train_test_split()</code> to split the data
- pass in the values into <code>mean_squared_error()</code>

The slides below provides a breakdown of the code below:

```python
from IPython.display import HTML
HTML("""<iframe src="https://docs.google.com/presentation/d/e/2PACX-1vRzlnAK6GgjxfDhWCe7-XhN_UPqwUSaU13cqMIDQOTnQHpphLvUwa9plhxsBKSiRA/embed?start=false&loop=false&delayms=60000" frameborder="0" width="990" height="585" allowfullscreen="true" mozallowfullscreen="true" webkitallowfullscreen="true"></iframe>""")
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features] # Print x out if you don't understand what data it contains
y = bank_advertising_df['Finance_Product_Sales']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Fit the model
linear_reg = linear_model.LinearRegression()
linear_reg_model = linear_reg.fit(X_train, y_train)

# Perform prediction with the model
y_pred = linear_reg_model.predict(X_test)

# Test the error
RMSE = mean_squared_error(y_test, y_pred, squared=False)

print(RMSE)
```

You can try printing <code>X_train, X_test, y_train, y_test</code> to understand the function of <code>train_test_split</code>

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

<b>3b.vii) Cross Validation (k-fold cross validation)</b>

Cross Validation is a is a resampling procedure used to evaluate machine learning models on a limited data sample. The procedure has a single parameter called k that refers to the number of groups that a given data sample is to be split into. As such, the procedure is often called k-foldcross-validation. 

To do so:

- we can use the <code>KFold()</code> to split the data
- iterate through each of the kfold
- calculate the <b>RMSE</b> for each of the fold, and append to a list

The slides below provides a breakdown of the code below:

```python
from IPython.display import HTML
HTML("""<iframe src="https://docs.google.com/presentation/d/e/2PACX-1vTTkt4wVuqetxuu0frRGd9xT0rWWvLAfYy9I1B3jpASYPs3QMAMfBzdGEkEUoHx8w/embed?start=false&loop=false&delayms=60000" frameborder="0" width="990" height="585" allowfullscreen="true" mozallowfullscreen="true" webkitallowfullscreen="true"></iframe>""")
```

```python
from sklearn.model_selection import KFold
import statistics

features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features]
y = bank_advertising_df['Finance_Product_Sales']

# Split the data into ten set
kf = KFold(n_splits=10)

# Create linear regression object
linear_reg = linear_model.LinearRegression()

# List of RMSE for each fold
k_fold_RMSE = []

# Iterate through each fold and calculate the RMSE for each fold
for train_index, test_index in kf.split(X):
    
    # Extract the training and test data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit model
    linear_reg_model = linear_reg.fit(X_train, y_train)
    y_pred = linear_reg_model.predict(X_test)
    
    #Calculate RMSE for the fold and append it
    RMSE = mean_squared_error(y_test, y_pred, squared=False)
    k_fold_RMSE.append(RMSE)
    
print('The RMSE for each fold is:', k_fold_RMSE)
print('The average RMSE is:', statistics.mean(k_fold_RMSE))
```

<div class="alert alert-block alert-warning">
Note that you should choose whether to use <b>train-test split</b> or <b>k-fold cross validation</b> based on your context:
    
- <b>train-test split</b> - Appropriate if computational resources is a constraint
- <b>k-fold cross validation</b> - This serves as a more holistic method of evaluation that makes more efficient use of your dataset. However, this is more computationally intensive than the train-test split method.

</div>

<b>3b.viii) Train model on the entire dataset</b>

Once we have determined the optimal (1) model, and (2) basket of features, that generates the best performance (based on RMSE), we are basically done with the research phase, and are now ready to operationalise the model for use by client. To do that, we proceed to <b>train the selected model on the selected basket of features on the entire dataset.</b>

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features]
y = bank_advertising_df['Finance_Product_Sales']

# Fit the model
linear_reg = linear_model.LinearRegression()
linear_reg_model = linear_reg.fit(X, y)
```

<div class="alert alert-block alert-warning">
Note that after training the model on the entire dataset, your model is complete. From here on, we can use this model <code>linear_reg_model</code> to generate predictions for new data. For example, let's say a client comes to you and state that they have a budget of 50 each on TV, Radio, and Newspaper, and ask you to estimate the effect it will have on their Sales, you can easily use your model to generate a prediction for them.

<b>4b.iii) Train Model</b>

Now we can begin to create the decision tree (regression). To do this, we make use of a library call - <b>scikit-learn (aka Sklearn)</b>. You can find Sklearn's documentations on linear regression in the link below:

- https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

We start by training the model on the features <b>TV</b>, <b>Radio</b> and <b>Newspaper</b>. To do so:
- We first create a decision tree regression object using the <code>DecisionTreeRegressor()</code> function
- On top of passing in the training dataset, We can pass in several default parameters:<br><br>
    - <code>max_depth</code> (<b>Used for pruning</b>. This determines the maximum depth of the tree)<br>
    - <code>max_leaf_nodes</code> (<b>Used for pruning</b>. This is the maximum number of leaf nodes in a tree.)<br>
    - <code>min_samples_leaf</code> (<b>Used for pruning but not advised unless you know what you are doing</b>. This sets the minimum number of samples required to form a leaf node.)<br>
    - <code>random_state</code> (simply sets a seed to the random generator, so that your train-test splits are always deterministic. If you don't set a seed, it is different each time.)<br><br>
- Thereafter we fit the model on the <b>training data</b> using the <code>fit()</code> function

```python
from sklearn.tree import DecisionTreeRegressor  
  
# create a regressor object 
regressor = DecisionTreeRegressor(max_depth=3, random_state=42)  
  
# Split data into predictors X and output Y
features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features] # Print x out if you don't understand what data it contains
y = bank_advertising_df['Finance_Product_Sales']

# fit the regressor with X and Y data
decision_tree_model = regressor.fit(X, y)
```

To see the structure of the tree that you built, you can do the following:

```python
from sklearn.tree import export_text

text_representation = export_text(regressor)
print(text_representation)
```

But if you want something even more visual, you could do the following:

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25,20))
_ = plot_tree(decision_tree_model, feature_names=['TV', 'Radio', 'Newspaper'], filled=True)
```

<b>4b.v) Calculate RMSE</b>

```python
from sklearn.metrics import mean_squared_error

actual_sales = [21.5] # actual sales is 21.5
RMSE = mean_squared_error(actual_sales, predicted_sales, squared=False) #squared is True by default

# Print the output
print('The RMSE of the prediction is:', RMSE)
```

<b>4b.vii) Validation by Measuring RMSE using Cross Validation (k-fold cross validation)</b>

To perform cross validation on decision tree (regression):

- we can use the <code>KFold()</code> to split the data
- iterate through each of the kfold
- calculate the <b>RMSE</b> for each of the fold, and append to a list

```python
from sklearn.tree import DecisionTreeRegressor  
from sklearn.model_selection import KFold
import statistics

features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features] # Print x out if you don't understand what data it contains
y = bank_advertising_df['Finance_Product_Sales']

# Split the data into ten set
kf = KFold(n_splits=10)

# Create linear regression object
regressor = DecisionTreeRegressor(random_state=42)  

# List of RMSE for each fold
k_fold_RMSE = []

# Iterate through each fold and calculate the RMSE for each fold
for train_index, test_index in kf.split(X):
    
    # Extract the training and test data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit model
    decision_tree_model = regressor.fit(X_train, y_train)
    y_pred = decision_tree_model.predict(X_test)
    
    #Calculate RMSE for the fold and append it
    RMSE = mean_squared_error(y_test, y_pred, squared=False)
    k_fold_RMSE.append(RMSE)
    
print('The RMSE for each fold is:', k_fold_RMSE)
print('The average RMSE is:', statistics.mean(k_fold_RMSE))
```

<div class="alert alert-block alert-warning">
Observe that the <b>RMSE of linear regression is higher than that of the decision tree</b>. This is likely because the features does not have a strong linear relationship with the outcome, which makes decision tree the more superior model in this case. This illustrates the point that training and testing a variety of model and then choosing the one with the best accuracy is a better approach than just going with our "favourite" model or our intuition.
</div>

4d. Pruning Decision Trees (Self-Study)
---

<div class="alert alert-block alert-success">
<b>4b.viii) Pruning (Optional)</b>

As covered earlier, in Sklearn, the tools you can use to prune the tree are through the following parameters of the <code>DecisionTreeRegressor</code> function:
- <code>max_depth</code>: The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
- <code>min_samples_split</code>: The minimum number of samples required to split an internal node
- <code>min_samples_leaf</code>: The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This may have the effect of smoothing the model, especially in regression.
</div>

Let's look at an example! But before we start, let's write k-fold cross validation into a method that takes in the following parameters:

- A model object (in this case we will pass in a decision tree model)
- The features of the dataset
- The outcome of the dataset

The method will return the following:
- The RMSE of the k-fold cross validation
- The number of leaves in the tree (obtained using the <code>get_n_leaves()</code> function)
- The depth of the tree (obtained using the <code>get_depth()</code> function)

```python
def k_fold_validation(regressor, X, y):
    
    # Split the data into ten set
    kf = KFold(n_splits=10)

    # List of RMSE for each fold
    k_fold_RMSE = []

    # Iterate through each fold and calculate the RMSE for each fold
    for train_index, test_index in kf.split(X):

        # Extract the training and test data
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # Fit model
        decision_tree_model = regressor.fit(X_train, y_train)
        y_pred = decision_tree_model.predict(X_test)
        
        # Get the number of leaves in the model
        num_of_leaves = decision_tree_model.get_n_leaves()
        
        # Get the depth of the tree
        depth_of_tree = decision_tree_model.get_depth()

        #Calculate RMSE for the fold and append it
        RMSE = mean_squared_error(y_test, y_pred, squared=False)
        k_fold_RMSE.append(RMSE)
    
    return depth_of_tree, num_of_leaves, statistics.mean(k_fold_RMSE)
```

A common approach to get the best possible tree is to (bear in mind this is a rather rudimentary way of pruning, and we have to resort to this since sklearn did not implement a proper method for pruning for regression tree):
- Grow a huge tree (for instance with <code>max_depth=8</code>)
- And slowly reduce the size of the tree
- Each time we reduce the size of the tree, we measure the RMSE of the tree using Cross-Validation
- Finally we take the tree with the lowest RMSE 

When we translate the above logic to pseudo code, it is as such:
- Iteratively create a model with a different <code>min_samples_split</code>
- Calculate the RMSE of that particular model
- Once we have tested enough models, look at the RMSE, and see which model yields the lowest RMSE. We will use this to determine the optimal <code>min_samples_split</code>

```python
from sklearn.tree import DecisionTreeRegressor  
from sklearn.model_selection import KFold
import statistics
from sklearn.metrics import mean_squared_error
import math

features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features] # Print x out if you don't understand what data it contains
y = bank_advertising_df['Finance_Product_Sales']
values= []
# Iteratively increase the number of samples required for each split (which limits the size of the tree)
for count in range(2, 200):

    # Create decision tree object
    regressor = DecisionTreeRegressor(random_state=42, max_depth=8, min_samples_split=count)
    
    # Calculate the kfold validation for the specific decision tree object
    depth_of_tree, num_of_leaves, rmse = k_fold_validation(regressor, X, y)
    
    # Formatting a print label
    label = 'Min Sample Split of ' + str(count) + ':'
    
    # Printing the output
    print(label, depth_of_tree, num_of_leaves, rmse)
    
    values.append(rmse)
```

<div class="alert alert-block alert-warning">
Observe that the RMSE is lowest (RMSE of 1.5595674587031543) when <code>min_sample_split=4</code>. This is indicative that <code>min_sample_split=4</code> is a viable setting in this case. <br>
    
Also take note you can see that when <code>min_sample_split=4</code>, the corresponding depth of the tree is <code>8</code> and the number of leaves is <code>63</code>
</div>

Finally, once you have determine the optimal parameter for the tree, we then use<code>min_sample_split=4</code> to build a decision tree, with which we can use to predict outcomes.

```python
# TV=292.9, Radio=28.3 and Newspaper=43.25
from sklearn.metrics import mean_squared_error

# Split data into predictors X and output Y
features = ['TV', 'Radio', 'Newspaper']
X = bank_advertising_df[features] # Print x out if you don't understand what data it contains
y = bank_advertising_df['Finance_Product_Sales']

# Features which we want to predict an outcome for
new_X = [[292.9, 28.3, 43.25]]  
actual_sales = [21.5] # actual sales for this set of feature is 21.5

# Create decision tree object
regressor = DecisionTreeRegressor(random_state=42, min_samples_split=4)

# fit the regressor with X and Y data 
decision_tree_model = regressor.fit(X, y) 
predicted_sales = decision_tree_model.predict(new_X)

# Calculate the error
RMSE = mean_squared_error(actual_sales, predicted_sales, squared=False) #squared is True by default

# Print the output
print('Predicted Sales:', predicted_sales[0])
print('The RMSE of the prediction is:', RMSE)
```
