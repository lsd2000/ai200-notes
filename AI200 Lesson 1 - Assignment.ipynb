Question 23: Pandas Multiple Conditional Select III [0.5 Mark]
---

<br>

<b>User story:</b>
- Display rows that fulfil the following conditions:
    - The <b>Property_Area</b> is "Semiurban" 
    - The <b>LoanAmount</b> is more than the average of all <b>loans</b> where the <b>Property_Area</b> is "Semiurban" 
- Hint: You can select part of the dataframe first with 1 of the conditions, store it in a temporary dataframe, and then apply the second condition on the temporary dataframe
- Expected output:<br><br>

<div style="margin-left:30px;background:white;">

<img src='
'/>    
    
</div>

```python
## Write your code below

mask = (loans_df["Property_Area"] == "Semiurban")
df_1 = loans_df.loc[mask]
mask2 = df_1["LoanAmount"] > df_1["LoanAmount"].mean()
df_2 = df_1.loc[mask2]

df_2

# Alternatively

m = loans_df["Property_Area"].eq("Semiurban")
mean_sem = loans_df.loc[m, "LoanAmount"].mean()

result = loans_df.loc[m & (loans_df["LoanAmount"] > mean_sem)]
result
```

```python
## Write your code below


def text_to_binary(state_str):
    if state_str == 'Y':
        return 1
    
    elif state_str == 'N':
        return 0
    
    return None
    
loans_df['Loan_Status'] = loans_df['Loan_Status'].apply(text_to_binary)
loans_df.head()

```
